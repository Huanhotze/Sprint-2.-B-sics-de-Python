## Exercici 3
Crea una llista amb nombres desordenats i respon a les següents preguntes:
Quants números hi ha?
- Quantes vegades apareix el número 3.
- Quantes vegades apareixen els nombres 3 i 4?
- Quin és el número més gran?
- Quins són els 3 números més petits?
- Quin és el rang d’aquesta llista?


```python
import random

num_list = [0] * 100
for i in range(100):
  num_list[i] = random.randint(0, 1000)

print(f"Quants números hi ha? {len(num_list)}" )
print(f"Quantes vegades apareix el número 3? {num_list.count(3)}" )
print(f"Quantes vegades apareixen els nombres 3 i 4? 3->{num_list.count(3)} 4->{num_list.count(4)} . Els numeros 3 y 4 en conjunt apareixen {num_list.count(3) + num_list.count(4)} vegades" )
print(f"Quin és el número més gran? {max(num_list)}")

num_list.sort()
m1, m2, m3, *others = num_list
print(f"Quins són els 3 números més petits? {m1}, {m2}, {m3}" )
print(f"Quin és el rang d’aquesta llista? Desde {num_list[0]} a {num_list[-1]}" )
```

## Exercici 4
Crea un diccionari de la següent forma i respon a les preguntes:
compra = { "Pomes" : {"Qty": 5, "€": 0.42}, "Peres" : {"Qty": 3, "€": 0.66} }

- Afegeix alguna fruita més
-     Quant han costat les peres en total?
-     Quantes fruites hem comprat en total?
-     Quina és la fruita més cara?


```python
compra = { "Pomes" : {"Qty": 5, "€": 0.42}, "Peres" : {"Qty": 3, "€": 0.66} }

#Afegeix alguna fruita més
compra["Platan"] = {"Qty": 15, "€": 0.99}

#Quant han costat les peres en total?
tot = compra["Peres"]["Qty"]*compra["Peres"]["€"]
print(f"Quant han costat les peres en total? {tot} €" )

#Quantes fruites hem comprat en total?
tot_qty = 0
for key in compra:
  tot_qty = tot_qty + compra[key]["Qty"]

print(f"Quantes fruites hem comprat en total? Un total de {tot_qty} unitats de {len(compra)} fruites diferents" )

#Quina és la fruita més cara?
max_preu = 0
for key, producte in compra.items():
  if producte["€"] > max_preu:
    max_preu = producte["€"]
    fruita = key

print(f"Quina és la fruita més cara? {fruita}")


```
