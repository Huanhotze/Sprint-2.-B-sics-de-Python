## Exercici 1
L'exercici consisteix a crear un programa que et classifiqui una variable numèrica en funció de l’escala Suspès/Aprovat/Notable/Excel·lent.

Recorda que Suspès < 5, Aprovat > 5 i < 7, Notable > 7 i < 9 i Excel·lent > 9.

```python
import random

num = round(random.random(), 3)*10

if num < 5:
  nota = "Suspes"
elif num >= 5 and num < 7:
  nota = "Aprovat"
elif num >= 7 and num < 9:
  nota = "Notable"
else:
  nota = "Excel.lent"

print(num)
nota
```

## Exercici 2
Utilitzant el següent tutorial crea un programa que et pregunti dos números. T’ha de mostrar un missatge dient si el primer és més gran, el segon és més gran o són iguals.


```python
num1 = input('Introduce un numero: ')
num2 = input('Introduce otro numero: ')

if num1 == num2:
  print("Ambos numeros son iguales")
elif num1 > num2:
  print(f"{num1} es mayor que {num2}")
else:
  print(f"{num2} es mayor que {num1}")
```

## Exercici 4
Crea un programa que donada una llista qualsevol, et digui si és simètrica o no. Si ho és, que et digui quants elements té.


```python
def simetrica(lista):
  lista2 = lista[::-1]
  if lista2 == lista:
    print(f"La lista {lista} es simetrica y tiene {len(lista)} elementos.")
  else:
    print(f"La lista {lista} no es simetrica")   



simetrica(['d', 'e', 'e', 'p', 'a', 'k'])
simetrica(['zero','one','two','three','four','five'])
simetrica(['zero','one','two','one','zero'])
simetrica([10,10,11,11,10,10])
```

## Exercici 5
Crea un programa que donada una llista, et digui quants números coincideixen amb la seva posició. Per exemple [3,4,2,0,2,3,6] el 2 i el 6 coincideixen.


```python
num_list = [0] * 10

for i in range(len(num_list)):
  num_list[i] = random.randint(0, 10)

count = 0
new_list = []
for i in range(len(num_list)):
  if i == num_list[i]:
    count = count + 1
    new_list.append(num_list[i])


print(num_list)
print(f"Coinciden {count} numeros. {new_list}")
```
